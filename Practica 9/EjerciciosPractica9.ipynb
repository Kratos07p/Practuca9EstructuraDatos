```python
class Articulo:

    def __init__(self, name, quantity, price) -> None:
        self.name = name
        self.quantity = quantity
        self.price = price
        
    def venta(self):
        print(f'Compraste {self.name} precio unitario {self.price} / {self.quantity} = Total {self.price * self.quantity}')



print('---Bienvenid@ a la Tiendita Programers---')
nombre = input('Ingresa el nombre del producto \n')
precio = float(input(f'Ingresa el precio de {nombre} \n'))
cantidad = int(input(f'Cuanta cantidad llevaria de {nombre} \n'))


leche = Articulo(nombre,cantidad,precio)
leche.venta()
```

```python
class Estudiante:


    def __init__(self , nombre, apellido , carnet, carrera) -> None:
        self.nombre = nombre
        self.apellido = apellido
        self.carnet = carnet
        self.carrera = carrera
        

    def mostrar_info (self):
        print(f'Nombre y Apellido del Alumno {self.nombre} {self.apellido}')
        print(f'Carnet de Alumno {self.carnet}')
        print(f'Carrera de Alumno {self.carrera}')
        
    def cambiar_nombre (self):
        self.nombre = input('Ingresa el nombre a actualizar del alumno\n')

    def cambiar_apellido (self):
        self.apellido = input('Ingresa el apellido a actualizar del alumno\n')

    def cambiar_carnet (self):
        self.carnet = input('Ingresa el carnet a actualizar del alumno\n')
    
    def cambiar_carrera (self):
        self.carrera = input('Ingresa la carrera a actualizar del alumno\n')
    
    

alumno = Estudiante('Pablo' , 'Gir칩n' , 'U20220407' , 'ING DS')

i=0
opcion = 0
while i == 0:
    alumno.mostrar_info()
    print('1 - Cambiar nombre')
    print('2 - Cambiar apellido')
    print('3 - Cambiar carnet')
    print('4 - Cambiar carrera')
    print('5 - Cerrar editor y guardar')

    opcion = int(input('Ingresa una opcion\n'))
    
    if opcion == 1:
        alumno.cambiar_nombre()
        print('--Dato Guaardado Correctamente--')
    elif opcion == 2:
        alumno.cambiar_apellido()
        print('--Dato Guaardado Correctamente--')
    elif opcion == 3:
        alumno.cambiar_carnet()
        print('--Dato Guaardado Correctamente--')
    elif opcion == 4:
        alumno.cambiar_carrera()
        print('--Dato Guaardado Correctamente--')
    elif opcion == 5:
        print('--Cerrando Programa--')
        i = 1
```
